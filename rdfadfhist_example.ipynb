# rdfadf example

Structural feature analysis for LiCoO2.  

```python
from rdfadf.rdf import compute_general_rdf
from rdfadf.adf import compute_general_adf

from ase.io import read
atoms = read('LiCoO2.cif', format='cif')
```

```python
r,g1,g1s,g2,g2s,cg,cg2=compute_general_rdf(atoms, "Li", "O", rcut=10.0, bins=100,sigma=1.0)
theta, adf, adfs = compute_general_adf(atoms, "d", "p", "p", rcut=3.0, bins=100, sigma=1.0)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(r, g2, label="g_raw / (4r)", marker='o')
plt.plot(r, g2s, label="g_smooth / (4r)", linewidth=2)

plt.xlabel("r ()")
plt.ylabel("g(r)")
plt.title("Gaussian-broadened RDF normalized by 4r")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(theta, adf, label="ADF raw", marker='o')
if adfs is not None:
    plt.plot(theta, adfs, label="ADF smoothed", linewidth=2)
plt.xlabel("Angle (rad)")
plt.ylabel("Counts")
plt.title("Angular Distribution Function (ADF)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
